```python
#import data from excel

import pandas as pd
df = pd.read_csv('DOAN_2020-08-11_Sentences.csv')
results_df = pd.DataFrame(df, columns = ['System', 'Sentence', 'Code'])

print(results_df.head(5))
```

```python
#pivot df
challenge_set = results_df.pivot('System', 'Sentence', 'Code')

#reorder systems
challenge_set.index = pd.CategoricalIndex(challenge_set.index, categories= ["Portage", "CNN", "RNN", "Attention"])
challenge_set.sort_index(level=0, inplace=True)

print(challenge_set.head(5))
```

```python
#create heatmap

import matplotlib.pyplot as plt
import seaborn as sns
challenge_set_transposed = challenge_set.transpose()

plt.figure(figsize=(15,30))

display_cs = sns.heatmap(challenge_set_transposed, center = 0, linewidths=.5, cmap="PiYG")
```
